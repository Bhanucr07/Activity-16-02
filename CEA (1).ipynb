```python
#importing required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("Bank_Personal_Loan_Modelling.xlsx",na_values=['.',':','?','!','@','#','$','%','^','&','*',','])
```

```python
df
```

```python
df.info()
```

```python
df = df.drop(["ID","ZIP Code",], axis=1)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.dtypes
```

```python
for col in ['Education','Securities Account','CD Account','Online','Personal Loan','Family','CreditCard']:
  df[col] = df[col].astype('category')
```

```python
df.dtypes
```

```python
cat_cols=['Family','Education','Personal Loan','Securities Account','CD Account','Online']
```

```python
num_cols=['Age','Income', 'Experience', 'CCAvg', 'Mortgage']
```

```python
df[num_cols].describe()
```

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('CreditCard',axis=1)
y= df['CreditCard']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=125, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
df.value_counts(['CreditCard'])/5000
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler = scaler.fit(X_train[num_cols])
X_train_num_std = pd.DataFrame(scaler.transform(X_train[num_cols]), columns=X_train[num_cols].columns)
X_test_num_std = pd.DataFrame(scaler.transform(X_test[num_cols]), columns=X_test[num_cols].columns)
```

```python
X_train_num_std.isnull().sum()
```

```python
X_test_num_std.isnull().sum()
```

```python
X_train_cat_dum = pd.get_dummies((X_train[cat_cols]),drop_first=True)
X_test_cat_dum = pd.get_dummies((X_test[cat_cols]),drop_first=True)
```

```python
X_train_cat_dum
```

```python
X_test_cat_dum
```

```python
X_train_num_std = X_train_num_std.reset_index()
X_train_num_std
```

```python
X_train_num_std = X_train_num_std.drop(["index"], axis=1)
```

```python
X_train_num_std
```

```python
X_train_cat_dum = X_train_cat_dum.reset_index()
X_train_cat_dum
```

```python
X_train_cat_dum = X_train_cat_dum.drop(["index"], axis=1)
```

```python
X_train_cat_dum
```

```python
X_test_num_std = X_test_num_std.reset_index()
X_test_num_std
```

```python
X_test_num_std = X_test_num_std.drop(["index"], axis=1)
```

```python
X_test_num_std
```

```python
X_test_cat_dum = X_test_cat_dum.reset_index()
X_test_cat_dum
```

```python
X_test_cat_dum = X_test_cat_dum.drop(["index"], axis=1)
```

```python
X_test_cat_dum
```

```python
final_X_train = pd.concat([X_train_num_std, X_train_cat_dum], axis=1)
```

```python
final_X_test = pd.concat([X_test_num_std, X_test_cat_dum], axis=1)
```

```python
final_X_train.isnull().sum()
```

```python
final_X_test.isnull().sum()
```

```python
final_X_train
```

```python
final_X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr = lr.fit(final_X_train,y_train)
```

```python
y_train_preds = lr.predict(final_X_train)
y_train_preds[0:10]
```

```python
y_test_preds = lr.predict(final_X_test)
y_test_preds[0:10]
```

```python
from sklearn.metrics import confusion_matrix
confusionmatrix = confusion_matrix(y_train,y_train_preds)
confusionmatrix
```

```python
confusionmatrix = confusion_matrix(y_test,y_test_preds)
confusionmatrix
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train,y_train_preds)
```

```python
accuracy_score(y_test,y_test_preds)
```
